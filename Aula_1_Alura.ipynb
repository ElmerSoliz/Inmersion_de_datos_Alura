```python
import pandas as pd

# Dando el valor de nuestros datos a una variable
inmuebles = pd.read_csv('/content/drive/MyDrive/Aula-1/inmuebles_bogota.csv')
# Usando head para traer solo 5 valores de la tabla
inmuebles.head()
```

```python
# Esto es para que nos traiga el numero de columnas y filas
inmuebles.shape
```

```python
# Para ver los atributos de columnas
inmuebles.columns
```

```python
# Creando un diccionario donde ponemos primero el valor de la columna del csv y luego el valor que lo reemplazara
columnas = {'Baños':'Banos','Área':'Area','Habitaciones':'Cuartos'}

# Renombrando las columnas que pusimos en el array
inmuebles = inmuebles.rename(columns = columnas)

# Usando sample para traer una cantidad determinada de datos pero de forma aleatoria
inmuebles.sample(10)
```

```python
# Nos muestra la informacion de la variable
inmuebles.info()
```

```python
# Me devuelve el valor del indice que le pacemos
inmuebles.iloc[300]
```

```python
# Estome devuelve en un rango de datos
inmuebles.iloc[300:305]
```

```python
# Esto me trae los valores dependiendo a la columna y tambien un rango
inmuebles['Valor'][300:305]
```

```python
# Esto me trae el tipo de dato
type(inmuebles['Valor'][300:305])
```

```python
# Esto devuelve el area promedio
inmuebles.Area.mean()
```

```python
# Esto lo que me trae son los valores que en barrio tenga Chico Reservado y me los trae como True
inmuebles.Barrio == 'Chico Reservado'
```

```python
# Esto suma todos los datos que tengan la variable Chico Reservado pero solo suma los que tenga true
sum(inmuebles.Barrio == 'Chico Reservado')
```

```python
inmuebles_chico = (inmuebles.Barrio == 'Chico Reservado')
inmuebles_chico
type(inmuebles_chico)
```

```python
# Estamos agragando los datos en un tipo array de inmuebles para poder mostrar los datos
chico_reservado = inmuebles[inmuebles_chico]
chico_reservado
```

```python
# Calculando el promedio del area de los 225 datos que teniamos
chico_reservado.Area.mean()
```

```python
# El numero total de barrios que hay sin contar los duplicados
len(inmuebles.Barrio.value_counts())
```

```python
# Esto me trae los barrios y suma la cntidad de repeticiones o barrios iguales que hay
inmuebles.Barrio.value_counts()
```

```python
# Usando el plot.bar para graficar y con ayuda del head ponemos un top 10
# tambien usamos el value_counts() para sacar los barrios pero sin duplicar es decir cuenta suma todos
inmuebles_barrio = inmuebles.Barrio.value_counts()
inmuebles_barrio.head(10).plot.bar()
```
